Step3: Create Procidure

```python
CREATE OR REPLACE PROCEDURE GetEmpName
IS
  employee SYS_REFCURSOR;  
BEGIN

  OPEN employee FOR 
  SELECT *
  from  employee;
  DBMS_SQL.RETURN_RESULT(employee);

END GetEmpName;
/
```

Step4: Run GetEmpName to showing result

```python
EXEC GetEmpName;
```
